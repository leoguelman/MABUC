### What if payouts are such that given intuition, we get different optimal actions based on the configuration of unobserved confounders?

So in the example below, if the intuition is to play machine M1, if D=0 and B=0, it would be optimal to play the counter-intuition; but if D=1 and B=1, it would be optimal to stick with the intuition. In this case, conditioning on intuition reflects the ETT by averaging out over all Unobvsered Confounders. In this case, although I cannot converge to the optimal policy, I still can do at least as good as or better using a Regret Decision Criterion. 

```python
m = Mabuc(pr_D=0.5, pr_B=0.5, T=1000, pr_payout = np.array([[0.1, 0.5, 0.4, 0.4],
                                                            [0.5, 0.1, 0.2, 0.2]]))
m.generate_samples(R=1000)
m.get_stats()
```

```python
m.plot()
```
